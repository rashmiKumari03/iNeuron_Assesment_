The data distributions like binominal distribution or possion distribution , they dont have the log normal or gaussian distribution.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(10)
num_of_trials = 10
probability = 0.6

bino_data = np.random.binomial(num_of_trials,probability,1000)
print(bino_data)
sns.displot(bino_data,kde='True')
plt.show()
```

```python
# For Continuous values we can talk power law distribution also..

# Parameters for the power-law distribution
alpha = 5  # Exponent
size = 1000  # Number of samples

# Generate power-law distributed data
data = np.random.power(alpha, size)

# Plot the histogram
plt.hist(data, bins=50, density=True, color='b')
plt.title(f'Power-law Distribution (={alpha})')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

Question: 3

What is the meaning of the five-number summary in Statistics? Give proper example

The five-number summary is a set of descriptive statistics that provides  summary of the distribution of a dataset. 
It consists of the following five values: Minimum value , Q1,Q2,Q3 and Maximum value...

* Minimum: The smallest observation in the dataset.
* First Quartile (Q1): The median of the lower half of the dataset.
* Median (Q2): The middle value of the dataset when it is ordered (Median).
* Third Quartile (Q3): The median of the upper half of the dataset.
* Maximum: The largest observation in the dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)
data = np.random.normal(size=100)  # Generating the normal distribution 

# Calculate the five-number summary
minimum = np.min(data)
q1 = np.percentile(data, 25)  # This will calculate the Q1 ie. 25% 
median = np.median(data)      # This will calculate Q2 ie. 50% 
q3 = np.percentile(data, 75)  # This will calclate Q3 ie. 75%
maximum = np.max(data)


print(f"Minimum: {minimum}")
print(f"First Quartile (Q1): {q1}")
print(f"Median (Q2): {median}")
print(f"Third Quartile (Q3): {q3}")
print(f"Maximum: {maximum}")
```

```python
# Plot the box plot
sns.boxplot(data)
plt.title('Box Plot with Five-Number Summary')
plt.ylabel('Values')
plt.show()
```

What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook

Correlation is a statistical measure that describes the extend that how much two variables are related to each other. 
The correlation coefficient quantifies the strength and direction of a linear relationship between two variables.Ranges from -1 to 1
* if r = 1 :  Perfect positive correlation (as one variable increases, the other also increases).
* if r = -1 : Perfect negative correlation (as one variable increases, the other decreases).
* if r = 0 :=0: No correlation (no linear relationship between the variables).

```python
# Loading the Iris dataset
iris = sns.load_dataset('iris')
iris.head()

# Numeric variables.
numeric_data = iris.select_dtypes(include=np.number)
categorical_data =iris.select_dtypes(include='object')
```
